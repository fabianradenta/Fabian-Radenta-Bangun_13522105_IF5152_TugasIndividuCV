```python
# imports
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import data
%matplotlib inline
```

```python
# import original image
ori_img = data.camera()
plt.imshow(ori_img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Gaussian Blurring

gaussian_kernel_size = (11,11)
gaussian_blurred = cv2.GaussianBlur(ori_img, gaussian_kernel_size, 0)

plt.imshow(gaussian_blurred, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.axis('off')
plt.show()
```

```python
# Median Blurring

median_kernel_size = 11
median_blurred = cv2.medianBlur(ori_img, median_kernel_size)

plt.imshow(median_blurred, cmap='gray')
plt.title('Median Blurred Image')
plt.axis('off')
plt.show()
```

```python
# comparison of filtering effects

plt.figure(figsize=(18,6))

# original image
plt.subplot(1,3,1)
plt.imshow(ori_img, cmap='gray')
plt.title('Original Image')
plt.axis('off') 

# Gaussian Blurred image
plt.subplot(1,3,2)
plt.imshow(gaussian_blurred, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.axis('off')

# Median Blurred image
plt.subplot(1,3,3)
plt.imshow(median_blurred, cmap='gray')
plt.title('Median Blurred Image')
plt.axis('off')

plt.suptitle('Comparison of Filtering Effects', fontsize=16)
plt.show()
```

```python
# saving images
cv2.imwrite('original_image.png', ori_img)
cv2.imwrite('gaussian_blurred_image.png', gaussian_blurred)
cv2.imwrite('median_blurred_image.png', median_blurred)

print("Images saved successfully.")
```
